```python
import cv2
import numpy as np
import mediapipe as mp
import torch
import torchvision.transforms as T
from torchvision.models.segmentation import deeplabv3_resnet50

# Загрузка моделей
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = deeplabv3_resnet50(pretrained=True).to(device).eval()

mp_pose = mp.solutions.pose
pose = mp_pose.Pose(model_complexity=2)
drawing = mp.solutions.drawing_utils

transform = T.Compose([
    T.ToPILImage(),
    T.Resize((520, 520)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225])
])

# Открытие видео
cap = cv2.VideoCapture('/content/drive/MyDrive/диплом/По элементам/Равновесия/Боковой шпагат с помощью, туловище в сторону в  горизонталь/Алина_Протасова равновесие планше — сделано в Clipchamp.mp4')
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video.mp4', fourcc, fps, (width, height))

frame_center = (width // 2, height // 2)  # Центр кадра

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # ======= 1. Сегментация через DeepLab =======
    input_tensor = transform(frame_rgb).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(input_tensor)['out'][0]
    output_predictions = output.argmax(0).byte().cpu().numpy()

    # В COCO класс 15 — человек
    person_mask = (output_predictions == 15).astype(np.uint8) * 255
    person_mask = cv2.resize(person_mask, (width, height), interpolation=cv2.INTER_NEAREST)

    # ======= 2. Поиск контуров =======
    contours, _ = cv2.findContours(person_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if contours:
        best_contour = None
        min_distance = float('inf')

        for contour in contours:
            area = cv2.contourArea(contour)
            if area < 500:  # Фильтрация маленьких шумов
                continue

            M = cv2.moments(contour)
            if M['m00'] == 0:
                continue
            cx = int(M['m10'] / M['m00'])
            cy = int(M['m01'] / M['m00'])

            distance = np.sqrt((cx - frame_center[0])**2 + (cy - frame_center[1])**2)

            if distance < min_distance:
                min_distance = distance
                best_contour = contour

        # Создание маски только для выбранного человека
        single_person_mask = np.zeros_like(person_mask)
        if best_contour is not None:
            cv2.drawContours(single_person_mask, [best_contour], -1, 255, thickness=cv2.FILLED)
    else:
        single_person_mask = np.zeros_like(person_mask)


    kernel = np.ones((5, 5), np.uint8)
    single_person_mask = cv2.morphologyEx(single_person_mask, cv2.MORPH_CLOSE, kernel)
    single_person_mask = cv2.morphologyEx(single_person_mask, cv2.MORPH_OPEN, kernel)
    single_person_mask = cv2.GaussianBlur(single_person_mask, (5, 5), 0)


    # ======= 3. Черный фон + силуэт =======
    black_frame = np.zeros_like(frame)
    silhouette_color = (180, 180, 180)  # Серый
    silhouette = black_frame.copy()
    silhouette[single_person_mask == 255] = silhouette_color

    # ======= 4. Рисуем суставы =======
    results_pose = pose.process(frame_rgb)
    if results_pose.pose_landmarks:
        drawing.draw_landmarks(
            silhouette,
            results_pose.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=drawing.DrawingSpec(color=(255, 255, 255), thickness=2, circle_radius=3),
            connection_drawing_spec=drawing.DrawingSpec(color=(200, 200, 200), thickness=2)
        )

    # ======= 5. Запись кадра =======
    out.write(silhouette)

cap.release()
out.release()

```

```python
import cv2
import numpy as np
import mediapipe as mp
import torch
import torchvision.transforms as T
from torchvision.models.segmentation import deeplabv3_resnet50
from ultralytics import YOLO

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

yolo_model = YOLO('yolov8s.pt')   # YOLOv8s
deeplab_model = deeplabv3_resnet50(pretrained=True).to(device).eval()

mp_pose = mp.solutions.pose
pose = mp_pose.Pose(model_complexity=2)
drawing = mp.solutions.drawing_utils

transform = T.Compose([
    T.ToPILImage(),
    T.Resize((520, 520)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225])
])

input_path = '/content/drive/MyDrive/диплом/По элементам/Повороты/Колбаса на полу/воротнева колбаса — сделано в Clipchamp.mp4'
# input_path = '/content/drive/MyDrive/диплом/По элементам/Равновесия/Затяжка с согнутой ногой/Алексия_Пашко равновесие затяжка с согнутой — сделано в Clipchamp.mp4'
# input_path = '/content/drive/MyDrive/диплом/По элементам/Повороты/Заднее/Анастасия_Иванова поворот заднее — сделано в Clipchamp.mp4'
cap = cv2.VideoCapture(input_path)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video_y.mp4', fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # 1. YOLO: Найти человека
    results = yolo_model.predict(frame, classes=[0], verbose=False)  # class 0 — person
    if results and results[0].boxes is not None and len(results[0].boxes) > 0:
        boxes = results[0].boxes.xyxy.cpu().numpy()

        # Выбираем самую большую область
        areas = [(x2 - x1) * (y2 - y1) for x1, y1, x2, y2 in boxes]
        main_box = boxes[np.argmax(areas)]
        x1, y1, x2, y2 = map(int, main_box)

        # Немного расширим рамку
        pad_w = int(0.1 * (x2 - x1))
        pad_h = int(0.1 * (y2 - y1))
        x1 = max(x1 - pad_w, 0)
        y1 = max(y1 - pad_h, 0)
        x2 = min(x2 + pad_w, width)
        y2 = min(y2 + pad_h, height)

    else:
        # Если нет обнаружений
        x1, y1, x2, y2 = 0, 0, width, height

    # 2. DeepLab: Сегментация внутри бокса
    cropped = frame_rgb[y1:y2, x1:x2]
    if cropped.size == 0:
        out.write(np.zeros_like(frame))
        continue

    input_tensor = transform(cropped).unsqueeze(0).to(device)
    with torch.no_grad():
        output = deeplab_model(input_tensor)['out'][0]
    output_predictions = output.argmax(0).byte().cpu().numpy()

    # Класс 15 = человек
    person_mask = (output_predictions == 15).astype(np.uint8) * 255

    # Маска обратно к оригинальному размеру бокса
    person_mask = cv2.resize(person_mask, (x2 - x1, y2 - y1), interpolation=cv2.INTER_NEAREST)

    # Полная черная маска кадра
    full_mask = np.zeros((height, width), dtype=np.uint8)
    full_mask[y1:y2, x1:x2] = person_mask

    # 3. Наложение силуэта на черный фон
    black_frame = np.zeros_like(frame)
    silhouette_color = (180, 180, 180)
    silhouette = black_frame.copy()
    silhouette[full_mask == 255] = silhouette_color

    # 4. MediaPipe: суставы на исходное видео
    results_pose = pose.process(frame_rgb)

    if results_pose.pose_landmarks:
        drawing.draw_landmarks(
            silhouette,
            results_pose.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=drawing.DrawingSpec(color=(255, 255, 255), thickness=2, circle_radius=3),
            connection_drawing_spec=drawing.DrawingSpec(color=(200, 200, 200), thickness=2)
        )

    # 5. Запись кадра
    out.write(silhouette)

cap.release()
out.release()
```
